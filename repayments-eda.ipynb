```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/kaggle/input/madedatasets/repayments.csv')
```

```python

# Display basic info
print("Dataset Overview:")
print(df.info())

```

```python
# Show summary statistics
print("\nSummary Statistics:")
print(df.describe())

```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

```

```python
# Univariate Analysis
plt.figure(figsize=(10,5))
df.hist(bins=30, figsize=(15,10), edgecolor='black')
plt.suptitle('Univariate Distribution of Numerical Features')
plt.show()

```

```python
# Multivariate Analysis
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Pairplot for selected numerical columns
sns.pairplot(df.select_dtypes(include=['number']))
plt.show()

```

```python
# Boxplots for categorical vs numerical relationships
categorical_cols = df.select_dtypes(include=['object']).columns
numerical_cols = df.select_dtypes(include=['number']).columns
for cat_col in categorical_cols:
    for num_col in numerical_cols:
        plt.figure(figsize=(10,5))
        sns.boxplot(x=df[cat_col], y=df[num_col])
        plt.xticks(rotation=45)
        plt.title(f'{num_col} by {cat_col}')
        plt.show()

print("EDA Completed!")
```
